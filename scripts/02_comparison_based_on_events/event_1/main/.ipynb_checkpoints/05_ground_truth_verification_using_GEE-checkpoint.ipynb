## Dataset Creation Using API Calls

```python
"NDVI":             # Normalized Difference Vegetation Index
"NDWI":             # Normalized Difference Water Index
"MNDWI":            # Modified Normalized Difference Water Index
"NDBI":             # Normalized Difference Built-up Index
"EVI":              # Enhanced Vegetation Index
"SAVI":             # Soil Adjusted Vegetation Index
"LAI":              # Leaf Area Index
"GCI":              # Green Chlorophyll Index
"ARVI":             # Atmospherically Resistant Vegetation Index
"BSI":              # Bare Soil Index
"LST_celsius":      # Land Surface Temperature (°C)
"canopy_cover_pct": # Canopy Cover Percentage (Hansen 2000 baseline)
```

```python

# ══════════════════════════════════════════════════════
#   ONLY EDIT THIS — then Run All
# ══════════════════════════════════════════════════════

SITE_ID  = "site_01"
LAT      = 22.65
LON      = 82.90

# ══════════════════════════════════════════════════════

GEE_PROJECT   = "ndvi-trend-481019"
OUTPUT_CSV    = "images_features.csv"
START_DATE    = "2013-01-01"
END_DATE      = "2026-01-01"
FREQ          = "monthly"
BUFFER_M      = 500
MAX_CLOUD_PCT = 20
L             = 0.5


def init_gee():
    ee.Initialize(project=GEE_PROJECT)
    print("✅ GEE initialised.")


def generate_date_ranges(start, end, freq):
    s = datetime.strptime(start, "%Y-%m-%d")
    e = datetime.strptime(end,   "%Y-%m-%d")
    delta = {"monthly": relativedelta(months=1),
             "quarterly": relativedelta(months=3),
             "yearly": relativedelta(years=1)}[freq]
    periods, cur = [], s
    while cur < e:
        nxt = min(cur + delta, e)
        label = cur.strftime(
            "%Y-%m" if freq == "monthly" else
            "%Y-Q" + str((cur.month-1)//3+1) if freq == "quarterly" else "%Y")
        periods.append((label, cur.strftime("%Y-%m-%d"), nxt.strftime("%Y-%m-%d")))
        cur = nxt
    return periods


def get_s2_image(region, start, end):
    return (ee.ImageCollection("COPERNICUS/S2_SR_HARMONIZED")
              .filterBounds(region).filterDate(start, end)
              .filter(ee.Filter.lt("CLOUDY_PIXEL_PERCENTAGE", MAX_CLOUD_PCT))
              .median().divide(10000))


def compute_indices(img):
    B2, B3, B4, B8, B11 = (img.select(b) for b in ["B2","B3","B4","B8","B11"])
    NDVI  = img.normalizedDifference(["B8","B4"]).rename("NDVI")
    NDWI  = img.normalizedDifference(["B3","B8"]).rename("NDWI")
    MNDWI = img.normalizedDifference(["B3","B11"]).rename("MNDWI")
    NDBI  = img.normalizedDifference(["B11","B8"]).rename("NDBI")
    EVI   = img.expression("2.5*((NIR-RED)/(NIR+6*RED-7.5*BLUE+1))",
                           {"NIR":B8,"RED":B4,"BLUE":B2}).rename("EVI")
    SAVI  = img.expression("((NIR-RED)/(NIR+RED+L))*(1+L)",
                           {"NIR":B8,"RED":B4,"L":L}).rename("SAVI")
    LAI   = EVI.multiply(3.618).subtract(0.118).rename("LAI")
    GCI   = B8.divide(B3).subtract(1).rename("GCI")
    ARVI  = img.expression("(NIR-(2*RED-BLUE))/(NIR+(2*RED-BLUE))",
                           {"NIR":B8,"RED":B4,"BLUE":B2}).rename("ARVI")
    BSI   = img.expression("((SWIR1+RED)-(NIR+BLUE))/((SWIR1+RED)+(NIR+BLUE))",
                           {"SWIR1":B11,"RED":B4,"NIR":B8,"BLUE":B2}).rename("BSI")
    return ee.Image([NDVI, NDWI, MNDWI, NDBI, EVI, SAVI, LAI, GCI, ARVI, BSI])


def get_lst(region, start, end):
    return (ee.ImageCollection("MODIS/061/MOD11A2")
              .filterBounds(region).filterDate(start, end)
              .select("LST_Day_1km").mean()
              .multiply(0.02).subtract(273.15).rename("LST_celsius"))


def get_canopy_cover(region):
    return (ee.Image("UMD/hansen/global_forest_change_2022_v1_10")
              .select("treecover2000").rename("canopy_cover_pct"))


def extract_point(lat, lon, period_label, s2_img, lst_img, canopy_img):
    buf  = ee.Geometry.Point([lon, lat]).buffer(BUFFER_M)
    try:
        vals = (s2_img.addBands(lst_img).addBands(canopy_img)
                .reduceRegion(reducer=ee.Reducer.mean(),
                              geometry=buf, scale=30, maxPixels=1e9)
                .getInfo())
    except Exception:
        vals = {}
    return {
        "id": SITE_ID, "lat": lat, "lon": lon, "period": period_label,
        "NDVI": vals.get("NDVI"),   "NDWI": vals.get("NDWI"),
        "MNDWI": vals.get("MNDWI"), "NDBI": vals.get("NDBI"),
        "EVI": vals.get("EVI"),     "SAVI": vals.get("SAVI"),
        "LAI": vals.get("LAI"),     "GCI": vals.get("GCI"),
        "ARVI": vals.get("ARVI"),   "BSI": vals.get("BSI"),
        "LST_celsius": vals.get("LST_celsius"),
        "canopy_cover_pct": vals.get("canopy_cover_pct"),
    }


def load_done_periods():
    """Skip periods already saved for this site."""
    if not os.path.exists(OUTPUT_CSV):
        return set()
    df = pd.read_csv(OUTPUT_CSV, usecols=["id", "period"], dtype={"id": str})
    return set(zip(df["id"], df["period"]))


# ── Run ───────────────────────────────────────────────────────
init_gee()

periods      = generate_date_ranges(START_DATE, END_DATE, FREQ)
done_periods = load_done_periods()
region       = ee.Geometry.BBox(82.50, 22.30, 83.30, 23.00)

print(f" Site: {SITE_ID}  ({LAT}, {LON})")
print(f"  {len(periods)} periods | already done: {len(done_periods)}\n")

records = []
for period_label, p_start, p_end in tqdm(periods, desc="Processing"):
    if (SITE_ID, period_label) in done_periods:
        continue   # skip if already in CSV
    s2_img     = compute_indices(get_s2_image(region, p_start, p_end))
    lst_img    = get_lst(region, p_start, p_end)
    canopy_img = get_canopy_cover(region)
    records.append(extract_point(LAT, LON, period_label, s2_img, lst_img, canopy_img))
    time.sleep(0.1)

if records:
    df = pd.DataFrame(records)
    write_header = not os.path.exists(OUTPUT_CSV)
    df.to_csv(OUTPUT_CSV, mode="a", index=False, header=write_header)
    print(f"\n✅ Appended {len(records)} rows to {OUTPUT_CSV}")
else:
    print("\n✅ Nothing new — all periods already in CSV.")

pd.read_csv(OUTPUT_CSV).tail(3)
```

# Visualisations

```python
# ╔══════════════════════════════════════════════════════════════════════════╗
# ║        HASDEO BUFFER — Data Scientist EDA Notebook                     ║
# ║        Temporal Feature Analysis · Correlation · Seasonality           ║
# ╚══════════════════════════════════════════════════════════════════════════╝
# Paste each cell block into a Jupyter notebook cell and run sequentially.
# Assumes df_cleaned is already loaded with columns:
#   id, lat, lon, period, NDVI, NDWI, MNDWI, NDBI, EVI, SAVI,
#   LAI, GCI, ARVI, BSI, LST_celsius, canopy_cover_pct

# ─────────────────────────────────────────────────────────────────────────────
# CELL 1 — Imports & Setup
# ─────────────────────────────────────────────────────────────────────────────
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns
from scipy import stats
from scipy.signal import savgol_filter

# ── Style ──────────────────────────────────────────────────────────────────
plt.style.use('dark_background')
DARK   = '#07100a'
SURF   = '#0e1a11'
GREEN  = '#39ff8a'
RED    = '#ff4f6e'
BLUE   = '#4dc9ff'
YELLOW = '#f5c44a'
DIM    = '#3d6649'
TEXT   = '#b8dfc0'

plt.rcParams.update({
    'figure.facecolor': DARK,
    'axes.facecolor':   SURF,
    'axes.edgecolor':   DIM,
    'axes.labelcolor':  TEXT,
    'xtick.color':      DIM,
    'ytick.color':      DIM,
    'text.color':       TEXT,
    'grid.color':       '#1c3022',
    'grid.linestyle':   '--',
    'grid.alpha':       0.5,
    'font.family':      'monospace',
    'axes.titlepad':    12,
})

FEATURES = ['NDVI','NDWI','MNDWI','NDBI','EVI','SAVI',
            'LAI','GCI','ARVI','BSI','LST_celsius','canopy_cover_pct']

VEG_FEATS   = ['NDVI','EVI','SAVI','LAI','GCI','ARVI']
WATER_FEATS = ['NDWI','MNDWI']
URBAN_FEATS = ['NDBI','BSI']
OTHER_FEATS = ['LST_celsius','canopy_cover_pct']

PALETTE = {
    'NDVI':'#39ff8a','NDWI':'#4dc9ff','MNDWI':'#00bcd4','NDBI':'#f5c44a',
    'EVI':'#76ff03','SAVI':'#b2ff59','LAI':'#00e676','GCI':'#69f0ae',
    'ARVI':'#40c4ff','BSI':'#ff6e40','LST_celsius':'#ff4f6e',
    'canopy_cover_pct':'#e040fb'
}

print("✅ Setup complete.")


# ─────────────────────────────────────────────────────────────────────────────
# CELL 2 — Load & Parse
# ─────────────────────────────────────────────────────────────────────────────

# Convert period to datetime and sort
df = df_cleaned.copy()
df['date'] = pd.to_datetime(df['period'])
df = df.sort_values('date').reset_index(drop=True)
df['year']  = df['date'].dt.year
df['month'] = df['date'].dt.month

print(f"Shape        : {df.shape}")
print(f"Period range : {df['date'].min().date()} → {df['date'].max().date()}")
print(f"Unique sites : {df['id'].nunique()}")
print(f"\nFeature stats:")
display(df[FEATURES].describe().round(4))


# ─────────────────────────────────────────────────────────────────────────────
# CELL 3 — Missing Values & Basic QC
# ─────────────────────────────────────────────────────────────────────────────

fig, axes = plt.subplots(1, 2, figsize=(14, 4), facecolor=DARK)
fig.suptitle('DATA QUALITY CHECK', fontsize=11, color=GREEN, fontfamily='monospace', y=1.02)

# Missing heatmap
miss = df[FEATURES].isnull().astype(int)
sns.heatmap(miss.T, ax=axes[0], cmap=['#0e1a11','#ff4f6e'],
            cbar=False, yticklabels=FEATURES, xticklabels=False)
axes[0].set_title('Missing Values (red = missing)', color=TEXT, fontsize=9)
axes[0].tick_params(labelsize=8)

# Feature value ranges
means = df[FEATURES].mean()
stds  = df[FEATURES].std()
y_pos = range(len(FEATURES))
axes[1].barh(y_pos, means, xerr=stds, color=[PALETTE[f] for f in FEATURES],
             alpha=0.8, height=0.6, ecolor=DIM, capsize=3)
axes[1].set_yticks(y_pos)
axes[1].set_yticklabels(FEATURES, fontsize=8)
axes[1].set_title('Feature Means ± Std Dev', color=TEXT, fontsize=9)
axes[1].axvline(0, color=DIM, linewidth=0.8)

plt.tight_layout()
plt.show()


# ─────────────────────────────────────────────────────────────────────────────
# CELL 4 — Temporal Trends: All Features Over Time
# ─────────────────────────────────────────────────────────────────────────────

fig, axes = plt.subplots(6, 2, figsize=(16, 20), facecolor=DARK)
fig.suptitle('TEMPORAL TRENDS — All Features (2018–2025)',
             fontsize=12, color=GREEN, fontfamily='monospace', y=1.005)
axes = axes.flatten()

for i, feat in enumerate(FEATURES):
    ax = axes[i]
    ts = df.groupby('date')[feat].mean()

    # Raw line
    ax.plot(ts.index, ts.values, color=PALETTE[feat], alpha=0.5,
            linewidth=0.8, label='Monthly mean')

    # Smoothed trend (Savitzky-Golay)
    if len(ts) >= 13:
        smooth = savgol_filter(ts.values, window_length=13, polyorder=2)
        ax.plot(ts.index, smooth, color=PALETTE[feat],
                linewidth=2, label='Smoothed trend')

    # Annual mean markers
    annual = df.groupby('year')[feat].mean()
    for yr, val in annual.items():
        ax.axhline(val, xmin=0, xmax=1, color=PALETTE[feat],
                   alpha=0.15, linewidth=0.5)

    ax.set_title(feat, color=PALETTE[feat], fontsize=9, fontfamily='monospace')
    ax.set_xlabel('', fontsize=7)
    ax.tick_params(labelsize=7)
    ax.legend(fontsize=6, loc='upper right')
    ax.grid(True)

plt.tight_layout()
plt.show()


# ─────────────────────────────────────────────────────────────────────────────
# CELL 5 — Seasonal Decomposition (Monthly Boxplots)
# ─────────────────────────────────────────────────────────────────────────────

fig, axes = plt.subplots(4, 3, figsize=(18, 14), facecolor=DARK)
fig.suptitle('SEASONALITY — Monthly Distribution per Feature',
             fontsize=12, color=GREEN, fontfamily='monospace')
axes = axes.flatten()
month_labels = ['J','F','M','A','M','J','J','A','S','O','N','D']

for i, feat in enumerate(FEATURES):
    ax = axes[i]
    monthly_data = [df[df['month'] == m][feat].dropna().values for m in range(1,13)]

    bp = ax.boxplot(monthly_data, patch_artist=True,
                    medianprops=dict(color=PALETTE[feat], linewidth=2),
                    flierprops=dict(marker='.', color=DIM, markersize=2),
                    whiskerprops=dict(color=DIM),
                    capprops=dict(color=DIM),
                    boxprops=dict(facecolor=PALETTE[feat]+'22', color=PALETTE[feat]))

    ax.set_xticks(range(1,13))
    ax.set_xticklabels(month_labels, fontsize=8)
    ax.set_title(feat, color=PALETTE[feat], fontsize=9)
    ax.tick_params(labelsize=7)
    ax.grid(True, axis='y')

plt.tight_layout()
plt.show()


# ─────────────────────────────────────────────────────────────────────────────
# CELL 6 — Correlation Heatmap (from your actual data)
# ─────────────────────────────────────────────────────────────────────────────

corr = df[FEATURES].corr()

fig, ax = plt.subplots(figsize=(13, 10), facecolor=DARK)
fig.suptitle('PEARSON CORRELATION MATRIX', fontsize=12,
             color=GREEN, fontfamily='monospace')

mask = np.zeros_like(corr, dtype=bool)  # show full matrix

cmap = sns.diverging_palette(0, 150, s=90, l=40, as_cmap=True)  # red→green

sns.heatmap(corr, ax=ax, mask=mask, cmap=cmap, center=0,
            vmin=-1, vmax=1, annot=True, fmt='.2f',
            annot_kws={'size': 7, 'color': 'white'},
            linewidths=0.5, linecolor='#1c3022',
            cbar_kws={'shrink': 0.8, 'label': 'Pearson r'})

ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right', fontsize=8)
ax.set_yticklabels(ax.get_yticklabels(), rotation=0, fontsize=8)
ax.set_title('Red = Negative  |  Green = Positive', color=DIM, fontsize=8, pad=10)

plt.tight_layout()
plt.show()

# Print top correlations
print("\n🔝 TOP 10 STRONGEST CORRELATIONS:")
corr_pairs = (corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))
                  .stack().reset_index())
corr_pairs.columns = ['Feature_A','Feature_B','Pearson_r']
corr_pairs['abs_r'] = corr_pairs['Pearson_r'].abs()
display(corr_pairs.sort_values('abs_r', ascending=False).head(10)
        .drop('abs_r', axis=1).reset_index(drop=True))


# ─────────────────────────────────────────────────────────────────────────────
# CELL 7 — Feature Group Trends (Vegetation vs Water vs Urban)
# ─────────────────────────────────────────────────────────────────────────────

fig, axes = plt.subplots(2, 2, figsize=(16, 10), facecolor=DARK)
fig.suptitle('FEATURE GROUPS — Temporal Evolution',
             fontsize=12, color=GREEN, fontfamily='monospace')

groups = {
    'Vegetation Indices': VEG_FEATS,
    'Water Indices':      WATER_FEATS,
    'Urban / Soil':       URBAN_FEATS,
    'LST & Canopy':       OTHER_FEATS,
}

for ax, (title, feats) in zip(axes.flatten(), groups.items()):
    for feat in feats:
        ts = df.groupby('date')[feat].mean()
        # Normalize for comparison
        ts_norm = (ts - ts.mean()) / ts.std()
        ax.plot(ts.index, ts_norm, label=feat,
                color=PALETTE[feat], linewidth=1.5, alpha=0.85)
    ax.set_title(f'{title} (Z-score normalised)', color=TEXT, fontsize=9)
    ax.legend(fontsize=8)
    ax.grid(True)
    ax.tick_params(labelsize=7)
    ax.axhline(0, color=DIM, linewidth=0.8, linestyle='--')

plt.tight_layout()
plt.show()


# ─────────────────────────────────────────────────────────────────────────────
# CELL 8 — Annual Trend Analysis (Year-over-Year)
# ─────────────────────────────────────────────────────────────────────────────

annual = df.groupby('year')[FEATURES].mean()

fig, axes = plt.subplots(3, 4, figsize=(18, 12), facecolor=DARK)
fig.suptitle('ANNUAL MEAN TRENDS — Year over Year',
             fontsize=12, color=GREEN, fontfamily='monospace')
axes = axes.flatten()

for i, feat in enumerate(FEATURES):
    ax = axes[i]
    vals = annual[feat]
    years = annual.index

    ax.bar(years, vals, color=PALETTE[feat], alpha=0.6, width=0.6)
    ax.plot(years, vals, 'o-', color=PALETTE[feat], linewidth=2, markersize=5)

    # Linear trend line
    slope, intercept, r, p, _ = stats.linregress(years, vals)
    trend = slope * years + intercept
    ax.plot(years, trend, '--', color='white', linewidth=1, alpha=0.5,
            label=f'trend r²={r**2:.2f} p={p:.3f}')

    direction = '▲' if slope > 0 else '▼'
    color_dir = GREEN if slope > 0 else RED
    ax.set_title(f'{feat}  {direction}', color=color_dir, fontsize=9)
    ax.legend(fontsize=6)
    ax.tick_params(labelsize=7)
    ax.grid(True, axis='y')
    ax.set_xticks(years)
    ax.set_xticklabels(years, rotation=45, fontsize=7)

plt.tight_layout()
plt.show()


# ─────────────────────────────────────────────────────────────────────────────
# CELL 9 — Scatter Matrix: Key Feature Pairs
# ─────────────────────────────────────────────────────────────────────────────

key_feats = ['NDVI','LST_celsius','NDWI','NDBI','EVI','BSI']
scatter_df = df[key_feats].dropna()

fig = plt.figure(figsize=(14, 12), facecolor=DARK)
fig.suptitle('SCATTER MATRIX — Key Feature Pairs',
             fontsize=12, color=GREEN, fontfamily='monospace')

n = len(key_feats)
for i, feat_y in enumerate(key_feats):
    for j, feat_x in enumerate(key_feats):
        ax = fig.add_subplot(n, n, i * n + j + 1)
        ax.set_facecolor(SURF)

        if i == j:
            # Diagonal: histogram
            ax.hist(scatter_df[feat_x], bins=30,
                    color=PALETTE[feat_x], alpha=0.8, edgecolor='none')
            ax.set_facecolor('#0a180d')
        else:
            r, p = stats.pearsonr(scatter_df[feat_x], scatter_df[feat_y])
            color = GREEN if r > 0 else RED
            ax.scatter(scatter_df[feat_x], scatter_df[feat_y],
                       alpha=0.3, s=3, color=color)
            # Regression line
            m, b = np.polyfit(scatter_df[feat_x], scatter_df[feat_y], 1)
            xs = np.linspace(scatter_df[feat_x].min(), scatter_df[feat_x].max(), 100)
            ax.plot(xs, m*xs+b, color='white', linewidth=1, alpha=0.7)
            ax.text(.05, .88, f'r={r:.2f}', transform=ax.transAxes,
                    fontsize=7, color=color)

        if i == n - 1: ax.set_xlabel(feat_x, fontsize=7, color=PALETTE[feat_x])
        if j == 0:     ax.set_ylabel(feat_y, fontsize=7, color=PALETTE[feat_y])
        ax.tick_params(labelsize=5)
        for spine in ax.spines.values():
            spine.set_edgecolor('#1c3022')

plt.tight_layout()
plt.show()


# ─────────────────────────────────────────────────────────────────────────────
# CELL 10 — Rolling Statistics (12-month window)
# ─────────────────────────────────────────────────────────────────────────────

key_temporal = ['NDVI', 'LST_celsius', 'NDWI', 'EVI']

fig, axes = plt.subplots(2, 2, figsize=(16, 10), facecolor=DARK)
fig.suptitle('ROLLING STATISTICS — 12-Month Window',
             fontsize=12, color=GREEN, fontfamily='monospace')

for ax, feat in zip(axes.flatten(), key_temporal):
    ts = df.groupby('date')[feat].mean()
    roll_mean = ts.rolling(12, center=True).mean()
    roll_std  = ts.rolling(12, center=True).std()

    ax.fill_between(ts.index,
                    roll_mean - roll_std,
                    roll_mean + roll_std,
                    color=PALETTE[feat], alpha=0.15, label='±1 Std Dev')
    ax.plot(ts.index, ts.values,
            color=PALETTE[feat], alpha=0.4, linewidth=0.8, label='Monthly')
    ax.plot(ts.index, roll_mean,
            color=PALETTE[feat], linewidth=2.5, label='12-mo rolling mean')

    ax.set_title(feat, color=PALETTE[feat], fontsize=10)
    ax.legend(fontsize=8)
    ax.grid(True)
    ax.tick_params(labelsize=8)

plt.tight_layout()
plt.show()


# ─────────────────────────────────────────────────────────────────────────────
# CELL 11 — Clustermap (Hierarchical Correlation Clustering)
# ─────────────────────────────────────────────────────────────────────────────

corr_clean = df[FEATURES].corr().dropna(axis=0, how='all').dropna(axis=1, how='all')

cg = sns.clustermap(
    corr_clean,
    cmap=sns.diverging_palette(0, 150, s=90, l=40, as_cmap=True),
    center=0, vmin=-1, vmax=1,
    annot=True, fmt='.2f',
    annot_kws={'size': 7},
    linewidths=0.5, linecolor='#1c3022',
    figsize=(12, 12),
    dendrogram_ratio=.15,
    cbar_pos=(1.02, 0.3, 0.03, 0.4)
)
cg.fig.set_facecolor(DARK)
cg.ax_heatmap.set_facecolor(SURF)
cg.fig.suptitle('HIERARCHICAL CLUSTERING — Feature Correlation',
                fontsize=12, color=GREEN, fontfamily='monospace', y=1.01)
plt.show()


# ─────────────────────────────────────────────────────────────────────────────
# CELL 12 — Summary Statistics Table
# ─────────────────────────────────────────────────────────────────────────────

summary = pd.DataFrame({
    'Mean':      df[FEATURES].mean(),
    'Std':       df[FEATURES].std(),
    'Min':       df[FEATURES].min(),
    'Max':       df[FEATURES].max(),
    'Skewness':  df[FEATURES].skew(),
    'Kurtosis':  df[FEATURES].kurtosis(),
}).round(4)

# Trend direction from linear regression
slopes = {}
for feat in FEATURES:
    ts = df.groupby('date')[feat].mean().dropna()
    if len(ts) > 2:
        slope, _, r, p, _ = stats.linregress(range(len(ts)), ts.values)
        slopes[feat] = {'Trend': '▲ Increasing' if slope > 0 else '▼ Decreasing',
                        'r²': round(r**2, 3), 'p-value': round(p, 4)}
    else:
        slopes[feat] = {'Trend': 'N/A', 'r²': None, 'p-value': None}

trend_df = pd.DataFrame(slopes).T
summary = pd.concat([summary, trend_df], axis=1)

print("\n📊 FEATURE SUMMARY TABLE:")
display(summary.style
    .background_gradient(subset=['Mean','Std'], cmap='Greens')
    .background_gradient(subset=['Skewness'], cmap='RdYlGn_r', vmin=-2, vmax=2)
    .format({'Mean':'{:.4f}','Std':'{:.4f}','Min':'{:.4f}',
             'Max':'{:.4f}','Skewness':'{:.3f}','Kurtosis':'{:.3f}'})
)
```
